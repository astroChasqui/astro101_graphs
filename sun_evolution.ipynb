```python
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import ascii
```

```python
# Data from http://people.sissa.it/~sbressan/CAF09_V1.2S_M36_LT/
# Z=0.017 track, M=1MSun
tr = ascii.read('data/track_sun.dat')
```

```python
plt.rc("font", family='serif', serif='Ubuntu', monospace='Ubuntu Mono', size=14)
plt.rc("axes", labelsize=15, titlesize=12)
plt.rc("xtick", top=True, direction='in', labelsize=14)
plt.rc("xtick.major", size=8, width=1)
plt.rc("ytick", right=True, direction='in', labelsize=14)
plt.rc("ytick.major", size=8, width=1)
plt.rc("lines", markersize=10, markeredgewidth=2)
plt.rc("lines", linewidth=3)

plt.figure(figsize=(6, 4.5))
plt.plot(tr['AGE']/1E9, 100*tr['H_CEN'], label='Hydrogen')
plt.plot(tr['AGE']/1E9, 100*tr['HE_CEN'], label='Helium')
plt.plot([4.6, 4.6], [0, 100], ls=':', color='green')
plt.text(3.2, 85, 'Today', color='green')
plt.legend(loc=5, borderaxespad=1.2)
plt.xlabel('Age of the Sun (billions of years)')
plt.title("Sun's core mass: Percentage of H or He", size=18, y=1.05)
plt.xlim(0, 10)
plt.ylim(0, 100)
plt.tight_layout()
plt.savefig('figures/SunCore_H_He.png')
```

```python
plt.figure(figsize=(6, 3))

plt.title("Sun's Luminosity ($L_\mathrm{Sun}$)", size=18, y=1.05)
plt.plot(tr['AGE']/1E9, 10**tr['LOG_L']*0.97)
plt.plot([4.6, 4.6], [0, 3], 'green', ls=':')
plt.plot([0, 10], [1.00, 1.00], 'green', ls=':')
plt.text(3.2, 1.75, 'Today', color='green')
plt.xlim(0, 10)
plt.ylim(0.6, 2.2)
plt.xlabel('Age of the Sun (billions of years)')
plt.tight_layout()
plt.savefig('figures/SunLuminosity.png')
```
